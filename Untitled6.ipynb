Numpy-
NumPy (Numerical Python) Python ki ek library hai jo arrays, mathematical operations, matrix operations, aur data analysis ke liye use hoti hai.

```python
import numpy as np
a = np.array([1,2,3,4])
print(a)
```

```python
arr=np.array([3,3,3,3.3,'hello',False])
print(bool(arr[-1])) # bool me agar value hoga toh wo true return karega , yaha "False" ek string hai jo ek value hai isliye true ayega

a= False
print(bool(a)) # yaha a ki value bool hai

b=0
print(bool(a))
```

```python
#np.zeros()- utme zeroz ka elemnt ka array banega ya toh matrix bhi bna sakte a = np.zeros(5)
print(a)
```

```python
b = np.zeros((2,3))
print(b)
```

```python
c = np.zeros((2,3,4))
print(c)
print(type(c[0][0][0]))
```

```python
d = np.ones((2,3,4),dtype=int)
print(d)
print(type(d[0][0][0]))
```

```python
#np.empty - Ye ek array banata hai without initializing value(garbage value )

e = np.empty(4,dtype = int)
print(e)
```

```python
#np.full - Ye poore array ko ek hi value se fill karta hai.

f= np.full(4,5,dtype=int)
print(f)
```

```python
import numpy as np

#np.arrange()- Ye start, stop, step ke basis par values generate karta hai.

a = np.arange(1, 10, 2)
print(a)
```

```python
#np.linspace()-Ye start aur stop ke beech exact number of values generate karta hai,yaha stop include hota hai

b=np.linspace(1, 100, 10)
print(b)

#np.linspace() mein endpoint=False ka matlab hai ki last value (stop value) include nahi hogi.

c = np.linspace(0, 10, 5, endpoint=False)
print(c)
```

```python
#shape - dimension of array
#size - total elemnets


a = np.array([[1,2,2],
              [3,4,4]])
print(a.size)
print(a.shape)
```

```python
#np.random.randint()- random value dega

import numpy as np

print(np.random.randint(1, 10))

arr = np.random.randint(1, 100, 5)
print(arr)
```

```python
#np.random.choice()- Ye kisi array/list se random element select karta hai.

import numpy as np

arr = [10, 20, 30, 40, 50]

print(np.random.choice(arr))
```
