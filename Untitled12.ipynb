```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("financial_loan.xlsx")
#display(df.head())
# df.info()
# df.describe()
#number_of_app = df['id'].nunique()
#print(number_of_app)

```

```python
curr_date = pd.Timestamp.today()
curr_month = curr_date.month
curr_day = curr_date.day

app = df[
    (df['issue_date'].dt.month == curr_month) &
    (df['issue_date'].dt.day <= curr_day)
]

total_loan_amount = app['loan_amount'].sum()

print('Loan amount:', total_loan_amount)
print(f'In millions :${total_loan_amount/1000000:.2f}M')
```

```python
loan_status_data = df.groupby('loan_status').agg({
    'id':'count',
    'dti':'mean',
    'int_rate':'mean',
    'loan_amount':'sum',
    'total_payment':'sum'
})

loan_status_data.rename(columns={'id':'Number of applications'},inplace = True)
display(loan_status_data)

```

## GOOD LOAN ANALYSIS

```python
#Good loan application
good_loan_apps = loan_status_data.iloc[1:3]['Number of applications'].sum()

print(good_loan_apps)
```

```python
#Good loan pplaication percentage
good_loan_percentage = (
    loan_status_data.iloc[1:3]['Number of applications'].sum()
    / loan_status_data['Number of applications'].sum()
) * 100

print(round(good_loan_percentage, 2), "%")
```

```python
#good loan funded amount
good_loan_funded = loan_status_data.iloc[1:3]['loan_amount'].sum()

print(round(good_loan_funded/1000000, 2), "M")
```

```python
# good loan total received
good_loan_received = loan_status_data.iloc[1:3]['total_payment'].sum()

print(round(good_loan_received/1000000, 2), "M")
```

```python
# Good loan application
good_loan_apps = (
    loan_status_data.loc['Fully Paid', 'Number of applications']
    + loan_status_data.loc['Current', 'Number of applications']
)

#Funded amount
good_loan_funded = (
    loan_status_data.loc['Fully Paid', 'loan_amount']
    + loan_status_data.loc['Current', 'loan_amount']
)

# Received amount
good_loan_received = (
    loan_status_data.loc['Fully Paid', 'total_payment']
    + loan_status_data.loc['Current', 'total_payment']
)

#percentage
good_loan_percentage = round(
    (good_loan_apps / loan_status_data['Number of applications'].sum()) * 100,
    2
)

#Profit
print("Profit:",good_loan_received - good_loan_funded)

print("Good Loan Applications:", good_loan_apps)
print("Good Loan Percentage:", good_loan_percentage, "%")
print("Good Loan Funded Amount:", round(good_loan_funded/1000000,2), "M")
print("Good Loan Received Amount:", round(good_loan_received/1000000,2), "M")
```

## BAD LOAN ANALYSIS

```python
# bad loan pplication
bad_loan_apps = loan_status_data.loc['Charged Off','Number of applications'].sum()
print(bad_loan_apps)

```

```python

# Applications
bad_loan_apps = loan_status_data.loc['Charged Off', 'Number of applications']

# Funded Amount
bad_loan_funded = loan_status_data.loc['Charged Off', 'loan_amount']

# Received Amount
bad_loan_received = loan_status_data.loc['Charged Off', 'total_payment']

# Percentage
bad_loan_percentage = round(
    (bad_loan_apps / loan_status_data['Number of applications'].sum()) * 100,
    2
)

print("Bad Loan Applications:", bad_loan_apps)
print("Bad Loan Percentage:", bad_loan_percentage, "%")
print("Bad Loan Funded Amount:", round(bad_loan_funded/1000000, 2), "M")
print("Bad Loan Received Amount:", round(bad_loan_received/1000000, 2), "M")
```

```python
monthly_loan = df.groupby(df['issue_date'].dt.month)['loan_amount'].agg('sum')
plt.plot(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],monthly_loan,marker='o',          # circle marker
    color='blue')

plt.tight_layout()
plt.title('Monthly Loan Amount')
plt.xlabel('Month')
plt.ylabel('Loan Amount')
plt.grid()

plt.show()
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun',
          'Jul','Aug','Sep','Oct','Nov','Dec']

monthly_loan = df.groupby(df['issue_date'].dt.month)['loan_amount'].sum()

monthly_received = df.groupby(df['issue_date'].dt.month)['total_payment'].sum()

plt.figure(figsize=(12,6))

plt.plot(
    months,
    monthly_loan/1000000,
    marker='o',
    color='blue',
    label='Loan Amount'
)

plt.plot(
    months,
    monthly_received/1000000,
    marker='o',
    color='green',
    label='Received Amount'
)

plt.title('Monthly Loan vs Received Amount')
plt.xlabel('Month')
plt.ylabel('Amount (Millions)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
