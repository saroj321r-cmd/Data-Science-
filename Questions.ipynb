```python
import numpy as np

n = int(input());
price =[];
for i in range(n):
  x= int(input())
  price.append(x)

grocery = np.array(price)

print("Grocery Prices:", grocery)
print("Total price:",np.sum(grocery))
print("Average:",np.mean(grocery))
print("Maximum Price:",np.max(grocery))
print("Minimum Price :",np.min(grocery))
print("Standard Deviation:",np.std(grocery))
```

```python
# create a 5*5 array(stock level)

stock = np.random.randint(1,90,(5,5))
print(stock)

# create another 5*5 (demand)

demand = np.random.randint(1,90,(5,5))
print(demand)

# remaining stock = stock - demand

remaining_stock = stock - demand
print(remaining_stock)

# neg value - 0 (out of stock)
outof_stock = np.where(remaining_stock<0,0,remaining_stock)
print(outof_stock)

# total shortage

total_shortage = np.sum(np.where(remaining_stock<0,remaining_stock* -1,0),axis =0)
print(total_shortage)

#  product with zero stock

zero_stock =np.sum(np.where(outof_stock==0,1,0),axis = 0)
print(zero_stock)

# sort each row and column on remaining stock

print(np.sort(remaining_stock,axis = 1))
print(np.sort(remaining_stock,axis=0))
```
