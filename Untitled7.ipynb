```python
import numpy as np

n = int(input("scores : "))

scores = []

for i in range(n):
    score = int(input())
    scores.append(score)

scores1 = np.array(scores)

print("Scores:", scores1)
print("Total Runs:", np.sum(scores1))
print("Average:", np.mean(scores1))
print("Highest Score:", np.max(scores1))
print("Lowest Score:", np.min(scores1))
print("Standard Deviation:", np.std(scores1))
print("Unique Scores:", np.unique(scores1))
```
