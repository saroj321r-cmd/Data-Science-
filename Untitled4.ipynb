List - mutuable

```python
lst = [2,3,0,-6,18]

#sort
lst.sort()
print(lst)

# sum
print(sum(lst))

#min
print(min(lst))

# max
print(max(lst))

# sorted - new list
print(sorted(lst))
```

```python
lst = [1, 2, "hello", 'c', [4, 5]]
print(lst)

#append()
lst.append([5, 6])
print(lst)

#extend()
lst.extend([5, 6])
print(lst)

#insert()
lst.insert(2, "saroj")
print(lst)
```

```python
lst = [1, 4, 6, 'a']
print(lst)

#pop()
lst.pop()
print(lst)

#remove()
lst.remove(4)
print(lst)

#index()
print(lst.index(6))

#count()
print(lst.count(4))
```
