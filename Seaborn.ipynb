```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('student_data.csv')

sns.lineplot(
    data=df,
    x='study_hours',
    y='exam_score'
)

plt.show()

```

```python
import pandas as pd
import seaborn as sns

def grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

df['grade']= df['exam_score'].apply(grade)

sns.barplot(data = df, x= 'grade',y= "exam_score",estimator = 'mean')
plt.title('performance')
plt.show()


```

```python
sns.scatterplot(
    data=df,
    x='study_hours',
    y='exam_score'
)

plt.show()
```

```python
sns.histplot(
    data=df,
    x='exam_score',
    bins=10,
    kde=True
)

plt.show()
```

```python
sns.boxplot(
    data=df,
    x='grade',
    y='exam_score'
)

plt.show()
```

```python
sns.violinplot(
    data=df,
    x='grade',
    y='exam_score'
)

plt.show()
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.show()
```
