```python
import pandas as pd

df1 = pd.DataFrame({
    'id':[1,2,3,4],
    'name':['JIN','JIHOO','THEV','JIMIN'],
    'age':[23,24,22,23]
})

df2 = pd.DataFrame({
    'id':[3,4,5,6],
    'name':['SUGA','Jhope','KOOKIE','RM'],
    'age':[23,24,22,23]
})

df3 = pd.merge(df1,df2,on = 'id',how = 'inner')  #inner join
print(df3)

df3.rename(columns={"name_x": "table_name1"},inplace = True)
print(df3)

df3 = pd.merge(df1,df2,on = 'id',how = 'left')  #left join
print(df3)

df3 = pd.merge(df1,df2,on = 'id',how = 'right')  #right join
print(df3)

df3 = pd.merge(df1,df2,on = 'id',how = 'outer')  #outer join
print(df3)

df3 = pd.merge(df1,df2,how = 'cross')  #cross join
print(df3)

df3 = pd.concat([df1,df2])
print(df3)


df4 = df1.drop_duplicates(subset = ['id'],keep = 'first')
df4

```
